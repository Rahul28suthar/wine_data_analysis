```python
quality_counts = df['quality'].value_counts()

plt.figure(figsize=[8,6])
plt.bar(quality_counts.index,quality_counts,color='pink')
plt.title("count plot of quality")
plt.xlabel("Quality")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=[8,6])
sns.swarmplot(x='quality',y='alcohol',data=df, palette = 'viridis')
plt.title("swarm plot for quality and alocohol")
plt.xlabel("quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
df['quality']= df['quality'].astype(str)

plt.figure(figsize=[10,8])
sns.violinplot(x='quality',y='alcohol',data=df, palette = 'viridis')
plt.title("violin plot for quality and alocohol")
plt.xlabel("quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
df['quality']= df['quality'].astype(str)

plt.figure(figsize=[10,8])
sns.boxplot(x='quality',y='alcohol',data=df)
plt.title("box plot for quality and alocohol")
plt.xlabel("quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
plt.figure(figsize=[15,10])
sns.heatmap(df.corr(),annot=True,fmt='.2f',cmap='Pastel2',linewidths=2)
plt.title("correlation heatmap")
plt.show()
```
